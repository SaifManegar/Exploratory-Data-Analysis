```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country
```

```python
final_data=pd.merge(df, df_country, on="Country Code", how="left")
final_data
```

```python
final_data.dtypes
```

```python
final_data.Country.value_counts()
```

```python
country_names=final_data.Country.value_counts().index
country_names
```

```python
country_values=final_data.Country.value_counts().values
country_values
```

```python
plt.pie(country_values[:4], labels=country_names[:4], autopct="%1.1f%%")
```

Observation: 93% of zomato records or transaction are happening from the country India

```python
final_data.columns
```

```python
final_data.groupby(["Aggregate rating", "Rating color", 'Rating text']).size()
```

```python
final_data.groupby(["Aggregate rating", "Rating color", 'Rating text']).size().reset_index()
```

```python
ratings=final_data.groupby(["Aggregate rating", "Rating color", 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

Observation: Whenever the ratings are below 2.4, it is rated poor & Whenever the ratings are between 3.0 to 3.4, it is rated average & Whenever the rating is above 4.5, it is rated excellent

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,4)
sns.set_style("darkgrid")
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings, hue="Rating color", palette=["white","red","orange","y","g","g"])
```

Observartion: 
*Not rated count is very high
*Maximmum number of ratings are between 2.6 to 4.2

```python
sns.countplot(x="Rating color", data=ratings, palette=["white","red","orange","y","g","g"])
```
